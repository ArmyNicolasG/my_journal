## Tensors
It's a vector or data point, but it can have any dimensional size. Each tensor has a different data type and size, like float32, int32, string and others. It's _shape_ represents the dimension of data.

### Creating a Tensor:

```python
import tensorflow as tf
string = tf.Variable("This is a string.", tf.string)
number = tf.Variable(324, tf.int64)
floating = tf.Variable(3.567, tf.float64)
```

### Rank/Degree of a Tensor:
The number of dimensions a Tensor has, the above created tensors are tensors of rank 0, also known as scalar.
These are some tensors of higher degree or ranks:

```python
rank1_tensor = tf.Variable(["More", "data", "in", "this", "one", "dimension", "tensor"], tf.string)
rank2_tensor = tf.Variable([ ["First dimension of this", "tensor"], ["Second dimension of", "this tensor"] ],tf.string)
```

**We use tensorflow.rank() method to know the rank** of rank2_tensor, which has a rank or degree of 2:

```python
tf.rank(rank2_tensor)
```

### Shape of a Tensor
We use **shape attribute** to get the shape of a Tensor, which represents the amount of elements that exist in each dimension of that Tensor. First parameter represents the amount of dimensions and second parameter how much values are there on each dimension (all dimensions must have the same size).

```python
rank2_tensor.shape
```

**Changing shape of a Tensor:**

```python
tensor1 = tf.ones([1,2,3]) # Creates a shape [1, 2, 3] Tensor full of ones.
tensor2 = tf.reshape(tensor1, [2,3,1]) # Reshape existing data to shape [2, 3, 1].
tensor3 = tf.reshape(tensor2, [3,-1]) # -1 tells the Tensor to calculate the size of the dimension in that place, this will reshape the tensor to [3, 3].

# The number of elements in the reshaped Tensor MUST match the number in the original.
```
